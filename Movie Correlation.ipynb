# Movie Correlation Project

In this project, we aim to determine what factors correlate directly in the overall success of the box office performance of a movie.

```python
# importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_csv(r'C:\Users\joefa\Downloads\archive(7)\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# See if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Remove rows with missing data

df = df.dropna()
```

```python
# Looking at datatypes

df.dtypes
```

```python
# Transforming columns to integers

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Creating a column which extracts the exact year from released column. Not all the original year values are accurate.

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
df.sort_values(by='gross', inplace=False, ascending=False)
```

An initial prediction we make when inspecting and thinking about the data, is that the budget of a movie has a direct correlation with the box office performance. We investigate this below.

```python
# Scatterplot with budget and gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# regression plot

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Investigate the correlations for all the numeric values

df.corr()
```

```python
# Visualise a correlation matrix

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Transform all columns into numeric values to investigate the correlations

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df
```

```python
# Check correlation matrix for significant values.

correlation_matrix2 = df_numerized.corr()

sns.heatmap(correlation_matrix2, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

# Conclusion

From the above data, we conclude that the 2 most significant values that correlate to the gross of a movie is the movie's budget and the amount of user votes rating the movie.
